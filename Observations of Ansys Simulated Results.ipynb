## The lesser the number of holes, the greater the CBL
## The smaller the web opening diameter, the greater the CBL

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
bucklingdata = pd.read_csv("CBL_SimulationResults.csv", encoding='cp1252')

# Extract relevant columns for analysis - Number of Holes
number_of_holes = bucklingdata['Number of holes, nh']
critical_buckling_load_nh = bucklingdata['Critical Buckling Load (kN)']

# Extract relevant columns for analysis - Web Opening Diameter
web_opening_diameter = bucklingdata['Web opening diameter, Do (mm)']
critical_buckling_load_do = bucklingdata['Critical Buckling Load (kN)']

# Extract relevant columns for analysis - Web-post width
web_post_width = bucklingdata['Web-post width, WP (mm)']
critical_buckling_load_do = bucklingdata['Critical Buckling Load (kN)']

# Set the figure size and create subplots
plt.figure(figsize=(8, 3))

# Plotting Boxplot for Number of Holes
plt.subplot(1, 3, 1)  # 1 row, 2 columns, subplot 1
boxplot_width = 0.7
plt.boxplot([critical_buckling_load_nh[number_of_holes == holes] for holes in number_of_holes.unique()],
            labels=number_of_holes.unique(),
            widths=boxplot_width,
            showfliers=False)  # Do not show outliers
plt.xlabel('Number of Holes, nh')
plt.ylabel('Critical Buckling Load (kN)')  # Y-label only in the first subplot
#plt.title('Effect of Number of Holes on Critical Buckling Load')
plt.grid(True)

# Plotting Boxplot for Web Opening Diameter
plt.subplot(1, 3, 2)  # 1 row, 2 columns, subplot 2
plt.boxplot([critical_buckling_load_do[web_opening_diameter == diameters] for diameters in web_opening_diameter.unique()],
            labels=web_opening_diameter.unique(),
            widths=boxplot_width,
            showfliers=False)  # Do not show outliers
plt.xlabel('Web Opening Diameter, Do (mm)')
#plt.title('Effect of Web Opening Diameter on Critical Buckling Load')
plt.grid(True)

# Plotting Boxplot for Web-post width
plt.subplot(1, 3, 3)  # 1 row, 3 columns, subplot 3
plt.boxplot([critical_buckling_load_do[web_post_width == widths] for widths in web_post_width.unique()],
            labels=web_post_width.unique(),
            widths=boxplot_width,
            showfliers=False)  # Do not show outliers
plt.xlabel('Web-post width, WP (mm)')
#plt.title('Effect of Web-post width on Critical Buckling Load')
plt.grid(True)
plt.tight_layout()
plt.show()
```
